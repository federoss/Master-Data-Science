b) Representa gráficamente la relación entre el importe de la factura y la propina. Genera un diagrama de dispersión utilizando la función __scatter__. Añade una tercera dimensión que será representada mediante distintos colores. La tercera dimensión se corresponde al sexo. Así podremos ver las facturas de los hombres y las de las mujeres. El sexo se corresponde con la variable en la posición 2 del array __tips__.

```python
# Sol b)
tercera_variable = tips[:,2]

fig, ax1 = plt.subplots(nrows = 1, figsize=(10,5))   
sc = ax1.scatter(variable1, variable2, c = tercera_variable, marker = 'o')  
ax1.set_title('Diagrama de dispersion con barra de colores')
ax1.set_xlabel('Variable 1')
ax1.set_ylabel('Variable 2')
plt.colorbar(sc)
```

```python
mi_fig, ax1 = plt.subplots(nrows= 1,
                           ncols = 1,
                           figsize=(10,5))


tips1[tips1.Sexo == 1].plot(kind='scatter', 
                            x='Precio_factura', 
                            y='Propina',
                            marker = 'D',
                            color='DarkBlue',
                            label = 'Hombres',
                            ax =  ax1);

tips1[tips1.Sexo == 2].plot(kind='scatter', 
                            x='Precio_factura', 
                            y='Propina',
                            marker = 'D',
                            color='DarkGreen',
                            ax = ax1,
                            label = 'Mujeres');

ax1.legend(bbox_to_anchor=(1.05, 1),
           loc=2, borderaxespad=0.,
           title = 'Leyenda');
```
